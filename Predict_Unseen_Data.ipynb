```python
import pickle
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

# Read Unseen Data

```python
df_test = pd.read_csv('test_data_finpro.csv')
df_test.head(5)
```

```python
df_test.drop(columns=["Effective To Date", "Customer", "Customer Lifetime Value"], inplace=True)
df_test.head(5)
```

# Load Model Random Forest

```python
with open('model/final_model_rf.pkl', 'rb') as file:
    loaded_model_rf = pickle.load(file)
```

# Load Model Kmeans 

```python
with open('model/final_model_kmeans.pkl', 'rb') as file:
    loaded_model_kmeans = pickle.load(file)
```

# Predict Regression

```python
predictions = loaded_model_rf.predict(df_test)
print(predictions)
```

```python
df_test['Customer Lifetime Value'] = predictions
df_test.head(5)
```

# Predict Clustering

```python
clusters = loaded_model_kmeans.fit_predict(df_test)
```

```python
df_test['Cluster'] = clusters
df_test.head(5)
```

```python
df_test.groupby("Cluster")["Customer Lifetime Value"].mean().sort_values(ascending=False)
```

The prediction of Customer Lifetime Value (CLV) using the Random Forest model and subsequent clustering analysis has provided valuable insights into customer segmentation. Based on the predictions, customers have been categorized into three distinct clusters:

- Gold Cluster (Cluster 2): This segment consists of high-value customers with significant profitability potential. It is crucial to implement strategies focused on retention and upselling premium services to this group.

- Silver Cluster (Cluster 0): This cluster includes moderate-value customers who present opportunities for growth. Tailored marketing efforts aimed at enhancing engagement can help elevate these customers to the Gold Cluster.

- Bronze Cluster (Cluster 1): Customers in this segment represent lower value, highlighting the need for targeted interventions to improve retention and increase their lifetime value.

This analysis emphasizes the importance of leveraging predictive modeling and clustering techniques to develop targeted marketing strategies, optimize resource allocation, and ultimately drive better customer engagement and profitability for CAV AutoInsurance.
